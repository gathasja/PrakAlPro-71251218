```python
#data yang diketahui
hargaBeli1 = 650000
beratEmas1 = 25
hargaSekarang = 685000

#keuntungan awal
modalAwal = beratEmas1*hargaBeli1
nilaiSekarang = beratEmas1*hargaSekarang
keuntunganRp = nilaiSekarang - modalAwal
keuntunganPersen = round((keuntunganRp/modalAwal)*100, 2)

print("Keuntungan awal Gerald ")
print("Keuntungan dalam Rupiah: Rp.", keuntunganRp)
print("Keuntungan dalam persen:", keuntunganPersen, "%")

#data yang diketahui pt.2
beratEmas2 = 15
totalEmas = beratEmas1 + beratEmas2

#total modal
modalKedua = beratEmas2*hargaSekarang
totalModal = modalAwal+modalKedua

#keuntungan kedua
hargaNaik = 715000
nilaiTotal = totalEmas*hargaNaik

keuntunganrp = nilaiTotal - totalModal
keuntunganpersen = round((keuntunganrp/totalModal)*100, 2)

print("\nKeuntungan kedua Gerald")
print("Keuntungan dalam Rupiah: Rp.", keuntunganrp)
print("Keuntungan dalam Persen:", keuntunganpersen, "%")


```
